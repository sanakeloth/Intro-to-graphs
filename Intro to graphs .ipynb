# Introduction to Graphs/Functions

```python
# In Python, lists and tuples are two different (and useful) ways to store groups of values.
simpleList = [1, 2, 3]
print(simpleList)
```

```python
# Lists are not limited to storing just numeric values. They can be used to store other types of data like Strings.

stringList = ['bread', 'and', 'butter']
print(stringList)
```

```python
# Try creating lists that store other types of data, like floating point numbers.
# Try creating a list that contains both numbers and Strings. What happened?

number = ["apple", 1, "cucumber"]
print(number)
```

```python
# In order to access an element of a list, we need to refer to its index. 
# Let's see how to do that with the lists we have created above.

a = simpleList[1]
b = simpleList[2]
print(a)
print(b)
```

```python
# You might find it strange that simpleList[1] equals 2 rather than 1. How did this happen?
# Well, the index of a list starts from 0-th position, not 1. 
# Below, try assign the first element of stringList to the label 'c'

c = stringList[0]
c
```

```python
# We can also start with an empty list, and then add elements to it.
# For example, below we create an empty list and then add a few values to it

emptyList = []
emptyList.append(1)
emptyList.append(2)
emptyList.append(3)
print(emptyList)
```

```python
# Next, we will see how to create a tuple in Python.
# The way to create a tuple looks quite similar to creating a list. Instead of enclosing the values in square brackets, 
# we enclose the values in a pair of parentheses.

simpleTuple = (1, 2, 3)
print(simpleTuple)
```

```python
# We can access the elements of a tuple in the same way as a list.

pos1 = simpleTuple[0]
pos2 = simpleTuple[1]
pos3 = simpleTuple[2]
print(pos1, pos2, pos3)
```

 Tuples behave pretty much like lists. However, once a tuple is created, the contents of the tuple cannot be changed.

```python
# We can use a for-loop to iterate over a list or a tuple.

simpleList = [1,2,3]
for val in simpleList:
    print(val)

simpleTuple = ['a','b','c']
for val in simpleTuple:
    print(val)
```

# Plotting Data with Matplotlib

**_Matplotlib_** is a Python package, which means that it is a collection of modules with related
functionality. 
* The modules in matplotlib are useful for plotting numbers and making graphs. 
* Matplotlib does not come built in with Pythonâ€™s standard library, before using it one has to  install it. However, if you have installed Anaconda, matplotlib will have been included. 

```python
%matplotlib inline
import matplotlib.pyplot as plt  # only need to import once
x_numbers = [1, 2, 3]
y_numbers = [2, 4, 6]
plt.plot(x_numbers, y_numbers)
plt.show()
```

```python
# You can mark points on the graph by passing an additional argument to the 'plot' method.
%matplotlib inline
x_numbers = [1, 2, 3]
y_numbers = [2, 4, 6]
plt.plot(x_numbers, y_numbers, 'o')
plt.show()
```

```python
# Let us turn to an example with more data points
%matplotlib inline
nyc_temp = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3]
plt.plot(nyc_temp, marker='*')
plt.show()
```

If you glance at the numbers we plotted, they really arenâ€™t very
far apart from each other. However, the graph makes the variations seem
rather dramatic. So, whatâ€™s going on? 

The reason is that matplotlib chooses
the range of the y-axis so that itâ€™s just enough to enclose the data supplied
for plotting. This makes even small differences look magnified because the range
of the y-axis is so small.

We will see below how we can customize the axes of a graph.

```python
# We can obtain the ranges of the current axes of a graph by calling the 'axis' function.

%matplotlib inline
nyc_temp = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3]
plt.plot(nyc_temp)
print(plt.axis())
plt.show()
```

axis() returns a tuple with four numbers, the first two for x-axis and the second two for the y-axis. You can cross-check with the axis of the graph plotted.

```python
# To change the range of either x-axis or y-axis, we use the same axis() method, 
# but this time we pass it named arguments to refer to the min and max value of the axis (or axes).

%matplotlib inline
nyc_temp = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3]
plt.plot(nyc_temp)
plt.axis(ymin=0)
plt.show()
```

The y-axis of the plot now starts at 0, and the differences between the values of the consecutive years look
less drastic.

```python
# Question? How can you plot more than one graph on the same plot?
# For example, if you want to compare the temperature records of New York City for three different years ...
# Please research and use the 'plot' method to achieve this below.

nyc_temp_2000 = [31.3, 37.3, 47.2, 51.0, 63.5, 71.3, 72.3, 72.7, 66.0, 57.0, 45.3, 31.1]
nyc_temp_2006 = [40.9, 35.7, 43.1, 55.7, 63.1, 71.0, 77.9, 75.8, 66.6, 56.2, 51.9, 43.6]
nyc_temp_2012 = [37.3, 40.9, 50.9, 54.8, 65.1, 71.0, 78.8, 76.7, 68.8, 58.0, 43.9, 41.5]
plt.plot(nyc_temp_2000)
plt.plot(nyc_temp_2006)
plt.plot(nyc_temp_2012)
plt.axis(ymin=0)
plt.legend(["nyc_temp_2000", "nyc_temp_2006", "nyc_temp_2012"])
plt.show()
```

A graph without a title, the axes titles, and a legend if you have more than one graph in the same plot won't be very informative. Can you explore the right methods in *__matplotlib__* to use to add these to the graph in the example so far. 


```python
# Write the Python code to add a title, the titles for x- and y-axis, and a legend to the plot of the example so far.
%matplotlib inline
import matplotlib.pyplot as plt  # only need to import once
x_numbers = [1, 2, 3]
y_numbers = [2, 4, 6]
plt.plot(x_numbers, y_numbers)
# displaying the title
plt.title("Linear graph")
# display titel for x and y
plt.xlabel('x_numbers') 
plt.ylabel('y_numbers') 
#display legend
plt.legend(["linear"])
plt.show()

```

# Exploring a Quadratic Function Visually

Let us consider the *__quadratic__* function $$y=x^2 + 2x + 1$$

For any value of *x*, the function produces some value for *y*.


Calculates the values of *y* for six different values of *x*. 

```python
'''
Quadratic function calculator
'''

# Assume values of x
x_values = [-1, 1, 2, 3, 4, 5]
for x in x_values:
    # Calculate the value of the quadratic function
    y = x**2 + 2*x + 1
    print('x={0} y={1}'.format(x,y))
```
